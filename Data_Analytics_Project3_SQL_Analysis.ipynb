## Explanation
Load the dataset and inspect its structure. This helps us understand columns, data types, and sample records before analysis.

```python
import pandas as pd
df = pd.read_excel('Dataset for Data Analytics (1).xlsx')
df.head()
```

## Explanation
Display only selected columns using the equivalent of a SQL SELECT query.

```python
df[['OrderID','CustomerID','Product','TotalPrice']].head()
```

## Explanation
Filter orders where TotalPrice is greater than 2000 (SQL WHERE clause).

```python
df[df['TotalPrice'] > 2000].head()
```

## Explanation
Filter orders paid using Credit Card and sort by TotalPrice in descending order (WHERE + ORDER BY).

```python
df[df['PaymentMethod']=='Credit Card'].sort_values('TotalPrice', ascending=False).head()
```

## Explanation
Count the total number of orders (COUNT aggregation).

```python
df['OrderID'].count()
```

## Explanation
Calculate total revenue generated from all orders (SUM aggregation).

```python
df['TotalPrice'].sum()
```

## Explanation
Calculate average order value (AVG aggregation).

```python
df['TotalPrice'].mean()
```

## Explanation
Group orders by Product and count the number of orders for each product (GROUP BY + COUNT).

```python
df.groupby('Product')['OrderID'].count().reset_index(name='OrderCount').sort_values('OrderCount', ascending=False)
```

## Explanation
Group orders by PaymentMethod and calculate total revenue for each payment type (GROUP BY + SUM).

```python
df.groupby('PaymentMethod')['TotalPrice'].sum().reset_index(name='Revenue').sort_values('Revenue', ascending=False)
```

## Explanation
Group orders by OrderStatus and calculate average order value (GROUP BY + AVG).

```python
df.groupby('OrderStatus')['TotalPrice'].mean().reset_index(name='AverageOrderValue')
```

## Explanation
Find the top 10 highest-value orders using ORDER BY.

```python
df.sort_values('TotalPrice', ascending=False).head(10)
```

## Explanation
Analyze referral sources by counting orders from each source.

```python
df.groupby('ReferralSource')['OrderID'].count().reset_index(name='Orders').sort_values('Orders', ascending=False)
```

## Explanation
Find total revenue generated by each referral source.

```python
df.groupby('ReferralSource')['TotalPrice'].sum().reset_index(name='Revenue').sort_values('Revenue', ascending=False)
```

## Explanation
Monthly revenue analysis using Date column.

```python
df['Date']=pd.to_datetime(df['Date'])
df.groupby(df['Date'].dt.to_period('M'))['TotalPrice'].sum().reset_index(name='Revenue')
```
